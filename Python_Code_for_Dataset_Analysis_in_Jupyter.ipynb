```python
age=int(input("Enter your age:"))
sys_bp=int(input("Enter your SystolicBP:"))
dias_bp=int(input("Enter your DiastolicBP:"))
blg=float(input("Enter your Blood Glucose level in mmol/L:"))
bd_temp=float(input("Enter your Body Temperature:"))
ht_rate=float(input("Enter your Heart Beat Rate:"))
```

```python
#age check
if((age>35)):
        flag_age=2
        print("Your age is more than the ideal age")
elif((age<=19)):
        flag_age=1
        print("Your age is less than the ideal age")
else:
        flag_age=0
        print("Your age is ideal")
#blood pressure check
if((sys_bp>120) or (dias_bp>80)):
        flag_bp=2
        print("Your blood pressure is high")
elif((sys_bp<110) or (dias_bp<70)):
        flag_bp=1
        print("Your blood pressure is low")
else:
        flag_bp=0
        print("Your blood pressure is normal")
#blood sugar check
if((blg>8.5)):
        flag_blg=2
        print("Your blood sugar is high")
elif(blg<4.5):
        flag_blg=1
        print("Your blood sugar is low")
else:
        flag_blg=0
        print("Your blood sugar is normal")
#heart rate check
if((ht_rate>85)):
        flag_ht=2
        print("Your heart rate is high")
elif(ht_rate<70):
        flag_ht=1
        print("Your heart rate is low")
else:
        flag_ht=0
        print("Your heart rate is normal")
#body temp check
if((bd_temp>99)):
        flag_bdt=2
        print("Your body temperature is high")
elif(bd_temp<96):
        flag_bdt=1
        print("Your body temperature is low")
else:
        flag_bdt=0
        print("Your body temperature is normal")
```

```python
if((flag_age==0)&(flag_bp==0)&(flag_blg==0)&(flag_ht==0)&(flag_bdt==0)):
    print("You are healthy and have low risk")
elif((flag_age==2)&(flag_bp==2)&(flag_blg==2)&(flag_ht==2)&(flag_bdt==2)):
    print("You have high risk")
elif((flag_age==1)&(flag_bp==1)&(flag_blg==1)&(flag_ht==1)&(flag_bdt==1)):
    print("You have high risk due to malnourishment or being underage")
else:
    print("You have mid risk")
```
